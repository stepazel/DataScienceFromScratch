
## Linear Regression Example - 2

1. Consider the `diabetes` dataset and illustrate the data points within a two-dimensional plot.

2. Use linear regression to draw a straight line that will best minimize the residual sum of squares between the observed responses in the dataset, and the responses predicted by the linear approximation.

3. Calculate the coefficients, residual sum of squares and the coefficient of determination.

(https://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html#sphx-glr-auto-examples-linear-model-plot-ols-py)


```python
# Code source: Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-200]
diabetes_X_test = diabetes_X[-200:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-200]
diabetes_y_test = diabetes_y[-200:]
```

```python
# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black");
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)

print(f"Intercept: {regr.intercept_}")
```

### MSE - Mean Squared Error

$$
MSE = \frac{\sum_{i}^{N}{(f(x_i) - \hat{f}(x_i))^2}}{N}
$$

- more information in 
https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html#sklearn.metrics.mean_squared_error


### $R^2$

Remember we define the residual sum of squares (RSS) as

$$
RSS=\sum_{i=1}^{n}e_i^2
$$
$$
R^2=1-\frac{RSS}{TSS}
$$

$R^2$ (coefficient of determination) regression score function.

- Best possible score is 1.0 and it can be negative. 



```python
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))


# The coefficient of determination: 1 is perfect prediction
# more information in https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score

print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))
```

### Plot outputs

```python
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
