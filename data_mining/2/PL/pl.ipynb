```python
# Count number of Nas per column
df.isnull().sum()
```

```python
# Remove line with Name, Team, Number, Position, Age, Height, Weight: NaN
df.dropna(subset=['Name', 'Team', 'Number', 'Position', 'Age', 'Height', 'Weight'])
```

```python
# Presents for each dateFrame column the number of NAs and their percentage
nulls = df.isnull().sum()
percentages = round(df.isnull().mean()*100, 1)
pd.concat([nulls, percentages], axis=1)
```

```python
# Fill in College column NaN values with 'no College'
# df.College = df.College.fillna('no College')
df.College.fillna('no College', inplace=True) #- this mutates the original DF
df
```

```python
df.drop(457, inplace=True)
```

```python
# View the lines with Salary NaN
df[df['Salary'].isna()]
```

```python
# Fill in the salary NaN values with the average salary value of the respective Team/Position
for row in df[df['Salary'].isna()].itertuples():
    df.at[row.Index, 'Salary'] = df[(df['Team'] == row.Team) & (df['Position'] == row.Position)]['Salary'].mean()
```

```python
# Count the number of Nas throughout all the dataframe
df.isna().sum().sum()
```
